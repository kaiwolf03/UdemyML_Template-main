```python
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from plotting import plot
```

```python
dataset = load_iris()

x = dataset.data[:, :2]
y = dataset.target
```

```python
colors = ["red", "green", "blue"]

for idx, point in enumerate(x):
    plt.scatter(point[0], point[1], color=colors[y[idx]])
plt.show()
```
